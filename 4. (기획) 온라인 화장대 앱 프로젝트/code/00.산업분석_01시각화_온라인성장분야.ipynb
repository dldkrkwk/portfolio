```python
import numpy as np
def sum_year(data,year,col_year):
    for i in range(len(data[data.columns[0]])):
        # year_total = 0
        # for x in range(col_year,col_year+12):
        #     yearly_total = data.iloc[i,x]
        yearly_total = np.sum(data.iloc[i,col_year:col_year+9])
        data.loc[i,year] = int(yearly_total/10000)
    return data
```

```python
type(data)
```

```python
data = sum_year(data,2020,2)
data = sum_year(data,2021,14)
data = sum_year(data,2022,26)
data = sum_year(data,2023,38)
data
```

```python
filtered = pd.concat([data.iloc[:,:2],data.iloc[:,-4:]],axis=1)
filtered
```

```python
total_index = [idx for idx,value in enumerate(filtered['운영형태']) if value == '계']
total_index
total_filtered = filtered.drop(total_index,axis=0)
total_filtered
```

```python

def calculate_yearly_rate(df):
    rate_df = df.copy()

    for year in range(2021, 2024):
        prev_year = year - 1
        current_year = year

        rate_df[current_year] = (df[current_year] - df[prev_year]) / df[prev_year] * 100

    rate_df[2020] = float('nan')

    rate_columns = ['대분류', '운영형태', 2020, 2021, 2022, 2023]
    rate_df = rate_df[rate_columns]

    return rate_df

yearly_rate_df = calculate_yearly_rate(total_filtered)
yearly_rate_df

```

```python
year_rate_df = yearly_rate_df.drop(2020,axis=1)
year_rate_df
```

```python
yearly_rate_df[yearly_rate_df['대분류']=='합계']
total = filtered[filtered['대분류']=='합계']
total = total.set_index('운영형태')
total = total.drop('대분류',axis=1)
total = total.T
total
```
