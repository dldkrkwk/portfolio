```python
df = pd.read_csv('./data/01.리뷰키워드분석_01통합본_올리브영_리뷰크롤링파일.csv')
df
```

```python
def keyword_check(df,keyword,brand):
    df.dropna(subset="content",axis=0,inplace=True)
    filtered_df = df[df['content'].str.contains(keyword)]
    filtered_df.to_excel(f'./data/01.리뷰키워드분석_02키워드분석_언급키워드확인_{brand}.xlsx',index=False)
    return filtered_df
```

```python
hawhae = keyword_check(df,'리뷰','화해')
hawhae = hawhae.sort_values('score',ascending=False)
hawhae
```

```python
# 무신사 뷰티 및 화장품 언급 수

df = pd.read_csv('./data/01.리뷰키워드분석_01통합본_무신사_리뷰크롤링파일.csv')
filtered_df = df[df['content'].str.contains('화장품|뷰티', regex=True)]
filtered_df
```

```python
def keyword_check(df,keyword,brand):
    df.dropna(subset="content",axis=0,inplace=True)
    filtered_df = df[df['content'].str.contains(keyword)]
    filtered_df.to_excel(f'./data/01.리뷰키워드분석_02키워드분석_언급키워드확인_{brand}.xlsx',index=False)
    return filtered_df
```

```python
df = pd.read_csv('./data/01.리뷰키워드분석_01통합본_에이블리_리뷰크롤링파일.csv')
```

```python
ably_keyword = keyword_check(df,'추천','에이블리')
ably_keyword = ably_keyword.sort_values('thumbsUpCount',ascending=False)
ably_keyword.head(30)
```
