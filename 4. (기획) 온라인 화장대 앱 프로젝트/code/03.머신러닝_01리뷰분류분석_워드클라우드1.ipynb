### 2) TF-IDF + kkma형태소
* kkma 형태소 적용(동사,명사,부사 뽑기)하고, 불용어 제거후/ tf-idf적용(가중치) 

```python
kkma = Kkma()
stop_words = '피부 하다 쓰다 안 바르다 사용 구매 없다 있다 따다 때 사용하다 리 다 같다 효과 사다 보다 느낌 느껴지다 제품 처음 정도' # 불용어

processed_docs=[]

for doc in results:
    morphs = kkma.pos(doc)
    picked_pos=[]

    for morph, pos in morphs:
        # 동사의 경우 '다'를 붙이고 불용어가 아닐 때만 추가
        if pos in ('VV', 'VA') and (morph + '다') not in stop_words:
            picked_pos.append(morph + '다')
        # 명사와 부사의 경우 불용어가 아닐 때만 추가
        elif pos in ('NNG', 'NNP', 'MAG') and morph not in stop_words:
            picked_pos.append(morph)
        
    processed_docs.append(" ".join(picked_pos))


# TfidfVectorizer 객체 생성 및 학습
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(processed_docs)

# 각 단어와 해당 단어의 tf-idf 값 출력
feature_names = vectorizer.get_feature_names_out()

for doc, vec in zip(results, tfidf_matrix.toarray()):
    print(f"Document: {doc}")
    print({feature_names[i]: vec[i] for i in range(len(feature_names))})
```
