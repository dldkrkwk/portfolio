## 01. 가설1 & 2 데이터 전처리

#### 01-01. 시군구명, 시군구코드에 스타벅스 매장 목록, 인구, 사업체 데이터를 추가

```python
# 서울시 스타벅스 매장 목록이 담긴 파일 불러오기

import pandas as pd

seoul_starbucks=pd.read_excel('./seoul_starbucks.xlsx',header=0)
seoul_starbucks.head()
```

```python
## 스타벅스 주소 data에서 '시군구명' 추출후, 필드로 추가

a=[]

for i in range(len(seoul_starbucks)):
    a.append(seoul_starbucks['주소'][i].split(' ')[1])

seoul_starbucks['시군구별'] = a

display(seoul_starbucks.head())
```

```python
## '시군구명'열 추가한 데이터, 엑셀로 저장 

seoul_starbucks.to_excel('../files/seoul_starbucks_list.xlsx', index=False)
```

### 01-2. 스타벅스 분석 데이터 만들기

* 시군구 목록 데이터 + 시군구별 스타벅스 매장수, 주민인구수, 사업체수

```python
import pandas as pd

## 시군구 목록데이터 불러오기

seoul_sgg = pd.read_excel('../files/seoul_sgg_list.xlsx')
seoul_sgg.head()
```

```python
## 스타벅스 매장목록 데이터 불러오기

seoul_starbucks_list = pd.read_excel('../files/seoul_starbucks_list.xlsx')
seoul_starbucks_list.head()
```

```python
## 시군구별 스타벅스 매장 수 세기

df1 = pd.pivot_table(seoul_starbucks_list,   # 피벗할 데이터프레임
                     index = '시군구별',    # 행 위치에 들어갈 열
                    #  columns = '',    # 열 위치에 들어갈 열
                     values = '매장명',     # 데이터로 사용할 열
                     aggfunc = 'count')   # 데이터 집계함수

df1.head(10)
```

```python
## 서울시 시군구 목록 데이터에 스타벅스 매장수 데이터를 병합

seoul_sgg.set_index('시군구명',inplace=True)
seoul_sgg=seoul_sgg.join(df1,how='left',sort=True)
```

```python
seoul_sgg.reset_index(inplace=True)
seoul_sgg.head()
```

```python
## 열이름 변경

seoul_sgg.rename(columns={'매장명':'스타벅스_매장수'},inplace=True)
seoul_sgg.head()
```

## 가설2

```python
## 시군구별 인구통계 엑셀 불러오기

seoul_sgg_pop= pd.read_excel('../files/sgg_pop.xlsx')
seoul_sgg_pop.head()
```

```python
## 서울시 시군구 목록 데이터에 서울시 시군구별 인구통계 데이터를 병합

seoul_sgg = pd.merge(seoul_sgg, seoul_sgg_pop, how='left', on ='시군구명')
seoul_sgg.head()
```

```python
## 서울시 종사자수,사업체수 엑셀 파일 불러오기

sgg_biz = pd.read_excel('../files/sgg_biz.xlsx')
sgg_biz.head(10)
```

```python
## 데이터프레임 병합

seoul_sgg = pd.merge(seoul_sgg, sgg_biz, how='left', on ='시군구명')
seoul_sgg.head()
```

```python
## 분석데이터 엑셀로 저장

seoul_sgg.to_excel('../files/seoul_sgg_stat.xlsx',index=False)
```

## 02. 가설 3 데이터 전처리

```python
import pandas as pd

## 원본 불러오기 

other_coffeeshops = pd.read_excel('../files/other_coffeeshops.xlsx')
other_coffeeshops.head()
```

```python
## 사용할 브랜드 3개만 따로 추출 (mega,ediya,twosome)

mega = other_coffeeshops[other_coffeeshops['브랜드']=='메가커피']
ediya= other_coffeeshops[other_coffeeshops['브랜드']=='이디야']
twosome = other_coffeeshops[other_coffeeshops['브랜드']=='투썸플레이스']
```

```python
## 결측치 확인
print("Mega 결측치 확인:")
print(mega.info(), end='\n')

print("\nEdiya 결측치 확인:")
print(ediya.info(), end='\n')

print("\nTwosome 결측치 확인:")
print(twosome.info(), end='\n')

# 메가- 좌표정보(9)
# 이디야 - 지번주소(1), 도로명주소(3), 좌표정보(7)
# 투썸플레이스 - 도로명주소(2), 좌표정보(8)
```

```python
## 결측치가 열마다 개수가 다른 데이터 살펴보기

# 이디야
print('이디야: 열별 결측치 data 확인\n')

df1 = ediya[ediya['지번주소'].isnull()]
df2 = ediya[ediya['도로명주소'].isnull()]
df3 = ediya[ediya['좌표정보(X)'].isnull()]
df4 = ediya[ediya['좌표정보(Y)'].isnull()]

display(df1,df2,df3,df4)

## 결론
# 8개 행의 결측치 (index_num: 943, 972, 1111, 1170, 1172, 1194, 1322, 1339)
```

```python
# 투썸플레이스
print('투썸플레이스: 열별 결측치 data 확인\n')

tdf1 = twosome[twosome['도로명주소'].isnull()]
tdf2 = twosome[twosome['좌표정보(X)'].isnull()]
tdf3 = twosome[twosome['좌표정보(Y)'].isnull()]

display(tdf1,tdf2,tdf3)

## 결론
# 10개 행의 결측치 (index_num: 522 ,625, 657, 685, 720, 726, 728, 764, 808, 820)
```

```python
## 결측치 제거

mega.dropna(inplace=True)
ediya.dropna(inplace=True)
twosome.dropna(inplace=True)
```

```python
## 중부원점 좌표계를 지도표시 가능한 위경도 좌표계로 변환

from pyproj import Proj, transform

# 객체저장(proj_1: 중부원점 좌표계, proj_2: 위경도 좌표계)
proj_1 = Proj(init='epsg:2097')
proj_2 = Proj(init='epsg:4326')

# 메가커피
converted = transform(proj_1, proj_2, mega['좌표정보(X)'].values, mega['좌표정보(Y)'].values)
mega['경도'] = converted[0]
mega['위도'] = converted[1]

# 이디야
converted2 = transform(proj_1, proj_2, ediya['좌표정보(X)'].values, ediya['좌표정보(Y)'].values)
ediya['경도'] = converted2[0]
ediya['위도'] = converted2[1]

# 투썸플레이스
converted3 = transform(proj_1, proj_2, twosome['좌표정보(X)'].values, twosome['좌표정보(Y)'].values)
twosome['경도'] = converted3[0]
twosome['위도'] = converted3[1]
```

```python
## '시군구별' 열 추가하기

mega['시군구별'] = mega['지번주소'].str.split(' ').str[1]
ediya['시군구별'] = ediya['지번주소'].str.split(' ').str[1]
twosome['시군구별'] = twosome['지번주소'].str.split(' ').str[1]
```

```python
## 쓸데없는 열 제거, 열 재배열

# 열제거
mega.drop(['지번주소','좌표정보(X)','좌표정보(Y)'],axis=1,inplace=True)
ediya.drop(['지번주소','좌표정보(X)','좌표정보(Y)'],axis=1,inplace=True)
twosome.drop(['지번주소','좌표정보(X)','좌표정보(Y)'],axis=1,inplace=True)

# 재배열
mega = mega[['브랜드','시군구별','도로명주소','사업장명','위도','경도']]
ediya = ediya[['브랜드','시군구별','도로명주소','사업장명','위도','경도']]
twosome = twosome[['브랜드','시군구별','도로명주소','사업장명','위도','경도']]
```

```python
## 엑셀로 저장 전 데이터 확인

mega.head()
```

```python
## 분석할 데이터, 각 엑셀 파일로 저장

mega.to_csv('mega.csv',index=False,encoding='euc-kr')
ediya.to_csv('ediya.csv',index=False,encoding='euc-kr')
twosome.to_csv('twosome.csv',index=False,encoding='euc-kr')
```
